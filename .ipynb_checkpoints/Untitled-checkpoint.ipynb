```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime
```

```python
ps = np.array([1])
npz = np.array([0,0])
inds = np.array([30])
N = 1
for t in range(8000):
  if t > 0 and t % 1000 == 0:
    ser = pd.Series(ps, index=inds) 
    #plt.plot(inds, ps, xlim=[-10, 110])
    ser.plot(figsize=[20,4], xlim=[-10, 110], title="{0}".format(t))    
    plt.show()
  inds = [inds[0]-1] + list(inds) + [inds[N-1]+1]
  ups = np.append(npz, ps * 0.5)
  downs = np.append(ps*.5, npz)
  #print(ups, downs)
  ps = ups + downs
  N = N+2
```

```python
ser = pd.Series(ps, index=inds)
print(ser[ser.index <= 0].sum())
print(ser[ser.index >= 100].sum())
print(ser[ser.index >= 30].sum())
print(ser.sum())
```
